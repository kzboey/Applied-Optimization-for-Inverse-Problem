Optional call back to save history

```python
fs = []
fy = []
step = 1
def callback1(x,y,i):
    if i % step == 0:
        fs.append(x)
        fy.append(y)
             
C = np.array([[2, 1], [1, 1]])
d = np.array([1, 1])
x0=np.array([15, 15])

```

```python
fs = []
fy = []
proxParams = {'v':d, 'x':x0, 'sigma':0.01, 'beta':0.1}
beta=proxParams['beta']

xmin, stopIdx = aomip.fast_pgm(C,d,x0,aomip.proximalL2Squared,proxParams,momentum=1,restart=True,restartItr=100,iteration=1000,callback=callback1)
print(xmin)
print(stopIdx)

plt.figure(figsize=(5,5))
axes = plt.gca()
fs_array = np.array(fs)
aomip.plotFunctionContour(aomip.myFunction, [-20, 20], 100, axes)
axes.plot(fs_array[:,0],fs_array[:,1], marker='.')
plt.show()
```
